```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = ""  # 나눔고딕 폰트 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 데이터 프레임 로드
accidentInfoList = pd.read_excel("./accidentInfoList.xlsx")

# '사고일시' 컬럼을 한글로 된 월과 시간 정보를 포함하는 형식으로 파싱
accidentInfoList['사고일시'] = pd.to_datetime(accidentInfoList['사고일시'], format='%Y년 %m월 %d일 %H시', errors='coerce')

# '사고일시' 컬럼을 날짜 형식으로 변환
accidentInfoList['사고일시'] = pd.to_datetime(accidentInfoList['사고일시'])

# '사고일시' 컬럼에서 월 정보 추출
accidentInfoList['월'] = accidentInfoList['사고일시'].dt.month

# 월별 사고 수 계산
월별_사고수 = accidentInfoList.groupby('월').size().reset_index(name='사고수')

# 막대 그래프로 시각화
plt.figure(figsize=(10, 6))
plt.bar(월별_사고수['월'], 월별_사고수['사고수'], color='skyblue')
plt.xlabel('월')
plt.ylabel('사고 수')
plt.title('월별 사고 수')
plt.xticks(range(1, 13))  # 월별로 X축에 표시
plt.show()
```

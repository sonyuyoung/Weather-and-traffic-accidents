```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "/System/Library/Fonts/AppleSDGothicNeo.ttc"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 데이터 프레임 로드 (이 부분을 실제 데이터에 맞게 수정해야 합니다.)
accidentInfoList = pd.read_excel("./accidentInfoList.xlsx")

# '시군구' 컬럼에서 '도'와 '시'를 추출하여 새로운 컬럼 생성
accidentInfoList['도'] = accidentInfoList['시군구'].apply(lambda x: x.split()[0])
accidentInfoList['시'] = accidentInfoList['시군구'].apply(lambda x: x.split()[1] if len(x.split()) > 1 else '')

# '도'를 기준으로 그룹화하여 사고 수 계산
도별_사고수 = accidentInfoList.groupby('도').size().reset_index(name='사고수')

# '시'를 기준으로 그룹화하여 사고 수 계산
시별_사고수 = accidentInfoList.groupby('시').size().reset_index(name='사고수')

# '면', '읍', '동'으로 끝나는 데이터 필터링
면읍동_끝_지역 = accidentInfoList[accidentInfoList['시군구'].str.endswith(('면', '읍', '동'))]

# '면읍동'을 기준으로 그룹화하여 사고 수 계산
면읍동_끝_사고수 = 면읍동_끝_지역.groupby('시군구').size().reset_index(name='사고수')

# 10건 이상인 경우 필터링
면읍동_끝_사고수 = 면읍동_끝_사고수[면읍동_끝_사고수['사고수'] >= 10]

# 도 별, 시 별, 면읍동 별 사고량 시각화
plt.figure(figsize=(15, 18))

# 도별 사고량
plt.subplot(4, 1, 1)
bars1 = plt.bar(도별_사고수['도'], 도별_사고수['사고수'], color='lightblue', edgecolor='black')  # 테두리 색 추가
plt.title('도 별 사고량')
plt.xlabel('도')
plt.ylabel('사고량')

# 각 막대 위에 사고수 표시
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 1), ha='center', va='bottom')

# 시 별 사고량
plt.subplot(4, 1, 2)
bars2 = plt.bar(시별_사고수['시'], 시별_사고수['사고수'], color='lightcoral', edgecolor='black')  # 테두리 색 추가
plt.title('시 별 사고량')
plt.xlabel('시')
plt.ylabel('사고량')

# 각 막대 위에 사고수 표시
for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 1), ha='center', va='bottom')

# 면읍동 별 사고량
plt.subplot(4, 1, 3)
bars3 = plt.bar(면읍동_끝_사고수['시군구'], 면읍동_끝_사고수['사고수'], color='lightgreen', edgecolor='black')  # 테두리 색 추가
plt.title('면읍동 별 사고량 (10건 이상)')
plt.xlabel('면읍동')
plt.ylabel('사고량')

# 각 막대 위에 사고수 표시 (텍스트 기울이기)
for bar, label in zip(bars3, 면읍동_끝_사고수['시군구']):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 1), ha='center', va='bottom', rotation=45)

# 사고일시별 사고량
plt.subplot(4, 1, 4)
accidentInfoList['사고일시'] = pd.to_datetime(accidentInfoList['사고일시'], format='%Y년 %m월 %d일 %H시', errors='coerce')

accidentInfoList['월'] = accidentInfoList['사고일시'].dt.month
월별_사고수 = accidentInfoList.groupby('월').size().reset_index(name='사고수')
bars4 = plt.bar(월별_사고수['월'], 월별_사고수['사고수'], color='lightpink', edgecolor='black')  # 테두리 색 추가
plt.title('월별 사고량')
plt.xlabel('월')
plt.ylabel('사고량')

# 각 막대 위에 사고수 표시
for bar in bars4:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 1), ha='center', va='bottom')

# x-axis labels rotation
plt.xticks(rotation=45, ha='right')

# '면읍동'별에서 지역이름 기울기 조절
plt.subplots_adjust(hspace=0.5)  # subplot 간격 조절
plt.show()
# 데이터프레임의 열 이름 확인
print(accidentInfoList.columns)

# '사고내용' 열에서 '경상사고', '중상사고' 등의 사고 종류를 뽑아내기
사고종류 = accidentInfoList['사고내용'].value_counts().reset_index(name='사고수')

# 결과 출력
print(사고종류)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "/System/Library/Fonts/AppleSDGothicNeo.ttc"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 데이터 프레임 로드 (이 부분을 실제 데이터에 맞게 수정해야 합니다.)
accidentInfoList = pd.read_excel("./accidentInfoList.xlsx")

# '시군구' 컬럼에서 '면', '읍', '동'으로 끝나는 데이터 필터링
면읍동_끝_지역 = accidentInfoList[accidentInfoList['시군구'].str.endswith(('면', '읍', '동'))]

# Convert the '가해운전자 상해정도' and '피해운전자 상해정도' columns to numeric
면읍동_끝_지역['가해운전자 상해정도'] = pd.to_numeric(면읍동_끝_지역['가해운전자 상해정도'], errors='coerce')
면읍동_끝_지역['피해운전자 상해정도'] = pd.to_numeric(면읍동_끝_지역['피해운전자 상해정도'], errors='coerce')

# 결측치를 평균값으로 대체
면읍동_끝_지역['가해운전자 상해정도'].fillna(면읍동_끝_지역['가해운전자 상해정도'].mean(), inplace=True)
면읍동_끝_지역['피해운전자 상해정도'].fillna(면읍동_끝_지역['피해운전자 상해정도'].mean(), inplace=True)

# '시군구'를 기준으로 그룹화하여 사고 수 및 가해운전자 상해정도, 피해운전자 상해정도 계산
지역별_사고상해 = 면읍동_끝_지역.groupby('시군구').agg({
    '가해운전자 상해정도': 'mean',
    '피해운전자 상해정도': 'mean',
    '사고번호': 'count'
}).reset_index()
지역별_사고상해.rename(columns={'사고번호': '사고수'}, inplace=True)

# 사고수 기준으로 내림차순 정렬 후 상위 5개 지역 선택
상위5지역 = 지역별_사고상해.sort_values(by='사고수', ascending=False).head(5)

# 가해운전자와 피해운전자 상해정도 추가
가해운전자_평균상해정도 = 면읍동_끝_지역.groupby('시군구')['가해운전자 상해정도'].mean().reset_index()
피해운전자_평균상해정도 = 면읍동_끝_지역.groupby('시군구')['피해운전자 상해정도'].mean().reset_index()

# 막대그래프 그리기
plt.figure(figsize=(15, 8))
plt.bar(상위5지역['시군구'], 상위5지역['사고수'], color='lightgreen', label='사고수')
plt.plot(상위5지역['시군구'], 상위5지역['가해운전자 상해정도'], marker='o', color='orange', label='가해운전자 상해정도')
plt.plot(상위5지역['시군구'], 상위5지역['피해운전자 상해정도'], marker='o', color='blue', label='피해운전자 상해정도')

# 가해운전자와 피해운전자 상해정도 선 그리기
for i, txt in enumerate(상위5지역['시군구']):
    plt.annotate(f'가해: {가해운전자_평균상해정도.iloc[i]["가해운전자 상해정도"]:.2f}\n피해: {피해운전자_평균상해정도.iloc[i]["피해운전자 상해정도"]:.2f}', (상위5지역['시군구'].iloc[i], 상위5지역['사고수'].iloc[i] + 5))

plt.title('끝의 면, 읍, 동으로 끝나는 지역의 사고량 및 상해정도 Top 5')
plt.xlabel('시군구')
plt.ylabel('건수 및 상해정도')
plt.legend()
plt.show()
```

```python
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# 휴게소 데이터 예시
rest_areas = pd.DataFrame({
    '휴게소명': ['서울만남(부산 방향)'],
    '주소': ['경기도 안성시 원곡면'],
    '위도': [37.460043],
    '경도': [127.042082]
})

# 사고 위치 데이터 예시
accidents = pd.DataFrame({
    '사고지': ['사고지1', '사고지2'],
    '주소': ['경기도 수원시 권선구', '경기도 용인시 기흥구'],
    '위도': [37.5, 37.6],
    '경도': [127.0, 127.1]
})

# 텍스트에서 '시'까지 추출하는 함수
def extract_city(address):
    city = address.split(' ')[1]  # 가정: 주소가 '도' 또는 '시'로 시작한다고 가정
    return city

# 휴게소와 사고 위치 간의 거리 계산 함수
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # approximate radius of Earth in kilometers
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2) ** 2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# 각 사고 위치에서 가장 가까운 휴게소 찾기
for i, accident in accidents.iterrows():
    min_distance = float('inf')
    nearest_rest_area = None
    
    for j, rest_area in rest_areas.iterrows():
        city_accident = extract_city(accident['주소'])
        city_rest_area = extract_city(rest_area['주소'])
        
        if city_accident == city_rest_area:
            distance = calculate_distance(accident['위도'], accident['경도'], rest_area['위도'], rest_area['경도'])
            
            if distance < min_distance:
                min_distance = distance
                nearest_rest_area = rest_area['휴게소명']
    
    print(f"사고 위치 '{accident['사고지']}'에서 가장 가까운 휴게소: {nearest_rest_area}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "/System/Library/Fonts/AppleSDGothicNeo.ttc"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 데이터 프레임 로드 (이 부분을 실제 데이터에 맞게 수정해야 합니다.)
accidentInfoList = pd.read_excel("./accidentInfoList.xlsx")

# '시군구' 컬럼에서 '면', '읍', '동'으로 끝나는 데이터 필터링
면읍동_끝_지역 = accidentInfoList[accidentInfoList['시군구'].str.endswith(('면', '읍', '동'))]

# '시군구'를 기준으로 그룹화하여 사고 수 계산
지역별_사고수 = 면읍동_끝_지역.groupby('시군구').size().reset_index(name='사고수')

# 사고수 기준으로 내림차순 정렬 후 상위 5개 지역 선택
상위5지역 = 지역별_사고수.sort_values(by='사고수', ascending=False).head(5)

# 막대그래프 그리기
plt.figure(figsize=(10, 6))
bars = plt.bar(상위5지역['시군구'], 상위5지역['사고수'], color='lightgreen')
plt.title('끝의 면, 읍, 동으로 끝나는 지역의 사고량 Top 5')
plt.xlabel('시군구')
plt.ylabel('사고량')

# 각 막대 위에 사고수 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 1), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "/System/Library/Fonts/AppleSDGothicNeo.ttc"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 데이터 프레임 로드 (이 부분을 실제 데이터에 맞게 수정해야 합니다.)
accidentInfoList = pd.read_excel("./accidentInfoList.xlsx")

# 'Si'로 시작하는 시군구 추출
시군구_Si = accidentInfoList[accidentInfoList['시군구'].str.startswith('Si')]

# 'Si'를 제외한 부분을 읍면동으로 나누기
시군구_Si['시'] = 시군구_Si['시군구'].apply(lambda x: x.replace('Si', '').strip())

# '시'를 기준으로 그룹화하여 사고 수 계산
지역별_사고수 = 시군구_Si.groupby('시').size().reset_index(name='사고수')

# 사고수 기준으로 내림차순 정렬 후 상위 5개 지역 선택
상위5지역 = 지역별_사고수.sort_values(by='사고수', ascending=False).head(5)

# 막대그래프 그리기
plt.figure(figsize=(10, 6))
bars = plt.bar(상위5지역['시'], 상위5지역['사고수'], color='lightgreen')
plt.title('시로 시작하는 지역의 사고량 Top 5 (읍면동 기준)')
plt.xlabel('읍면동')
plt.ylabel('사고량')

# 각 막대 위에 사고수 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 1), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 가상의 데이터 생성 (이 부분을 실제 데이터에 맞게 수정해야 합니다)
data = {'시군구': ['*면', '*읍', '*동'],
        '사고건수': [5, 3, 7, 2, 4]}
df = pd.DataFrame(data)

# '면, 읍, 동'으로 끝나는 지역명의 데이터 필터링
면읍동_끝_지역 = df[df['시군구'].str.endswith(('면', '읍', '동'))]

# '시군구'를 기준으로 그룹화하여 사고 수 계산
지역별_사고수 = 면읍동_끝_지역.groupby('시군구')['사고건수'].sum().sort_values(ascending=False)

# 상위 5개 지역 선택
상위5지역 = 지역별_사고수.head(5)

# 막대그래프 그리기
plt.figure(figsize=(10, 6))
상위5지역.plot(kind='bar', color='lightgreen')
plt.title('끝의 면, 읍, 동으로 끝나는 지역의 사고량 Top 5')
plt.xlabel('시군구')
plt.ylabel('사고량')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
font_path = "/System/Library/Fonts/AppleSDGothicNeo.ttc"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# CSV 파일 경로
file_path = '경부고속도로_전체_accidentInfoList.csv'

# CSV 파일 읽기 (encoding 지정)
df = pd.read_csv(file_path, delimiter=',', encoding='cp949')

# '사고일시' 컬럼을 datetime 형식으로 변환
df['사고일시'] = pd.to_datetime(df['사고일시'], format='%Y년 %m월 %d일 %H시')

# '월' 컬럼 추가
df['월'] = df['사고일시'].dt.month

# 월별 사고 수 계산
monthly_accidents = df.groupby('월').size().reset_index(name='사고수')

# '기상상태' 컬럼이 있다고 가정
# '기상상태'에 따른 월별 사고 수 계산
weather_monthly_accidents = df.groupby(['월', '기상상태']).size().reset_index(name='사고수')

# 막대 그래프 그리기
plt.figure(figsize=(12, 8))

# 월별 사고 수 그래프
sns.barplot(x='월', y='사고수', data=monthly_accidents, color='skyblue', label='월별 사고 수')

# 날씨에 따른 월별 사고 수 그래프
sns.barplot(x='월', y='사고수', hue='기상상태', data=weather_monthly_accidents)

plt.title('월별 및 날씨에 따른 사고 수')
plt.xlabel('월')
plt.ylabel('사고 수')
plt.xticks(rotation=0)
plt.legend(title='기상상태', loc='upper right', bbox_to_anchor=(1.2, 1))

plt.show()
```

```python
import folium
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Haversine formula to calculate distance between two points on the Earth
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # approximate radius of Earth in kilometers
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2) ** 2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# 첫 번째 CSV 파일 로드
CB_geoData = pd.read_csv("./accidentInfoList_x_y.csv")

# 특정 위치를 중심으로 하는 맵 생성
map_CB = folium.Map(location=[37.51467818712183, 127.032052458702], zoom_start=14)

# 빨간색 마커로 첫 번째 CSV 파일에 대한 마커 추가
for i, store in CB_geoData.iterrows():
    popup_text = f"사고 발생 시간: {store['field2']}"  # 원하는 정보로 수정
    folium.Marker(location=[store["_Y"], store["_X"]],
                  icon=folium.Icon(color="red", icon="star"),
                  popup=popup_text).add_to(map_CB)

# 경기도에 해당하는 행에 동그라미 마커 추가
gyeonggi_data = CB_geoData[CB_geoData['field4'].str.contains('경기도')]
if not gyeonggi_data.empty:
    latitudes = gyeonggi_data["_Y"].tolist()
    longitudes = gyeonggi_data["_X"].tolist()
    lat_center = sum(latitudes) / len(latitudes)
    lon_center = sum(longitudes) / len(longitudes)
    max_distance = max([haversine(lat_center, lon_center, lat, lon) for lat, lon in zip(latitudes, longitudes)])
    
    folium.Circle(location=[lat_center, lon_center],
                  radius=max_distance * 1000 * 1.05,  # Convert km to meters and add 5% margin
                  color='blue',  # 테두리 색상
                  fill=False).add_to(map_CB)

# 두 번째 CSV 파일 로드
RestCsv_data = pd.read_csv("./RestCsv.csv")

# 파란색 마커로 두 번째 CSV 파일에 대한 마커 추가
for i, store in RestCsv_data.iterrows():
    popup_text = f"휴게소: {store['휴게소명']}"  # 원하는 정보로 수정
    folium.Marker(location=[store["위도"], store["경도"]],
                  icon=folium.Icon(color="blue", icon="star"),
                  popup=popup_text).add_to(map_CB)

# 세 번째 CSV 파일 로드
Rest2Csv_data = pd.read_csv("./경부선졸음쉼터표준데이터.csv")

# 초록색 마커로 세 번째 CSV 파일에 대한 마커 추가
for i, store in Rest2Csv_data.iterrows():
    popup_text = f"cctv: {store['졸음쉼터명']}"  # 원하는 정보로 수정
    folium.Marker(location=[store["위도"], store["경도"]],
                  icon=folium.Icon(color="green", icon="star"),
                  popup=popup_text).add_to(map_CB)

# 네 번째 CSV 파일 로드
cctvCsv_data = pd.read_csv("./cctv.csv")

# 분홍색 마커로 네 번째 CSV 파일에 대한 마커 추가
for i, store in cctvCsv_data.iterrows():
    popup_text = f"CCTV 설치 위치: {store['field3']}"  # 원하는 정보로 수정
    folium.Marker(location=[store["_Y"], store["_X"]],
                  icon=folium.Icon(color="pink", icon="star"),
                  popup=popup_text).add_to(map_CB)

# 맵을 HTML 파일로 저장
map_CB.save("./교통사고마커8.html")
```
